## Multivariate Analysis

```python
df.columns
df_corr = df[['popularity', 'duration_ms', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature']]
df_corr
```

```python
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(df_corr.corr(), annot=True, cmap='mako')
```

```python
sns.pairplot(df[['acousticness','energy']], height=5, aspect=1)
```

```python
df_sample = df[['acousticness','energy']].sample(n=5000, random_state=1)
sns.pairplot(df_sample, height=5, aspect=1.5, diag_kind='kde')
```

## Add more variables in pair plot

```python
df['track_genre'].value_counts()

df_genres = df[df['track_genre'].isin(['pop','folk','emo','punk-rock','electronic'])]

df_sample = df_genres.sample(n=5000, random_state=1)

sns.pairplot(df_sample[['acousticness','energy','track_genre']], height=5, aspect=1.5, diag_kind='kde', hue='track_genre')
```
